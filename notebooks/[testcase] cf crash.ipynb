```python
from glob import glob
from datetime import datetime

import numpy
import cf

data_dir = '/data/cmip5/output1/MOHC/HadGEM2-ES/rcp85/mon/atmos/' \
           'Amon/r1i1p1/v20111215/tas/'
fs = glob(data_dir + 'tas_Amon_HadGEM2-ES_rcp85_r1i1p1_*-*.nc')
t0 = datetime(2010, 1, 1)
t1 = datetime(2010, 2, 1)

def get_time_range_cf (time):
    f = cf.Coordinate()
    f._data = cf.Data(numpy.array([0, 0]))
    u = cf.Units('months since 2010-1-1')
    u.calendar = time.calendar
    f.override_Units(u)
    f.slice[:] = [0, 1]
    f.units = time.units
    return f.array

def get_data_cf ():
    fields = cf.read(fs)
    datasets = []
    for f in fields:
        if f.name() == 'air_temperature':
            s = f.space
            time = s['dim0']
            if time.size == 0:
                continue
            else:
                t_1d = time.size == 1
            this_t0, this_t1 = get_time_range_cf(time)# = date2num(total_t0, time.units, time.calendar)
            #t1 = date2num(total_t1, time.units, time.calendar)
            time = time.varray
            if t_1d:
                time = [time]
            if time[-1] >= this_t0 and time[0] < this_t1:
                # add data for each time in the right range
                lat = s['dim1']
                lon = s['dim2']
                for i_t, t in enumerate(time):
                    if this_t0 <= t < this_t1:
                        tas = f.slice[:,:].array if t_1d else f.slice[i_t,:,:].array[0]
                        datasets.append((t, lon.array, lat.array, tas))
    cf.close()
    return datasets

get_data_cf()
```
